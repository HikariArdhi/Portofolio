```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='DarkRed')
        
    fig.tight_layout()  
    plt.show()
draw_histograms(data,data.columns,10,5)
```

```python
# Data numerik
datax = data[data["label2"] == 0]
dataxx = datax[:200000]

# Data kategorikal
datay = data[data["label2"] == 1]

neww = pd.concat([dataxx, datay], ignore_index=True,sort=False)
neww
```

```python
from sklearn.utils import shuffle
new = shuffle(neww)
new
```

```python
X = new.iloc[:,:-1].values
y = new.iloc[:,-1].values
```

```python
from imblearn.over_sampling import SMOTE
X, y = SMOTE().fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)

scaler2 = MinMaxScaler()
X_train = scaler2.fit_transform(X_train)
X_test = scaler2.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE

rfe = RFE(RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1), n_features_to_select = 15) #15 feature yang dipilih (bisa diganti tergantung evaluasi)
rfe.fit(X_train, y_train)
rfe.get_support()
print("Random Forest accuracy  :",rfe.score(X_test,y_test))
```

```python
y_pred =rfe.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),
                    y_test.reshape(len(y_test),1)),1))
```

```python
hasil_predict = pd.DataFrame({'Actual' : y_test, 'Predict' : y_pred})
hasil_predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report


cm = confusion_matrix(y_test,y_pred)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual :0', 'Actual :1'], 
                                 index=['Predict :0', 'Predict :1'])

print(classification_report(y_test,y_pred))
print("\n")
print("Training set score for Recursive feature extraction (RFE): %f" % rfe.score(X_train , y_train))
print("Testing  set score for Recursive feature extraction (RFE): %f" % rfe.score(X_test  , y_test ))
print("\n")
print("Confusion Matrix :")
sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```
